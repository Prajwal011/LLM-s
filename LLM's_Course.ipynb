# **Introduction to Large Language Models (LLMs)**

# **What are LLMs?**

LLMs, or large language models, are a type of artificial intelligence (AI) that are trained on massive datasets of text and code. They can be used to perform a variety of tasks, such as natural language generation, natural language translation, question answering, and text summarization.

LLMs are made up of a large number of interconnected neurons, which are trained to recognize patterns in the data. This allows them to learn the statistical relationships between words and phrases, and to generate new text that is similar to the text they were trained on.


## **Definition and Background of LLMs**

Large Language Models (LLMs) refer to a class of artificial intelligence models designed to understand and generate human language. These models are built using deep learning techniques, particularly transformer-based architectures, to process natural language data. LLMs have significantly impacted the field of Natural Language Processing (NLP) and have demonstrated remarkable capabilities in various language-related tasks.

The primary goal of LLMs is to learn the underlying patterns and structures within large corpora of text data and use that knowledge to generate coherent and contextually relevant text. By employing sophisticated algorithms, these models can handle vast amounts of data and exhibit a form of "language understanding" that was previously thought to be beyond the capabilities of machines.

**Applications and Use Cases of LLMs**

LLMs have found applications in various domains and have proven to be powerful tools for solving a wide range of language-related tasks. Some prominent use cases include:

1. **Language Generation:** LLMs can generate coherent and contextually relevant text, making them useful for applications like creative writing, storytelling, and chatbots.

2. **Machine Translation:** Transformers have shown remarkable performance in machine translation tasks, helping bridge the language barrier between different cultures.

3. **Question Answering:** LLMs can answer questions based on the context provided, making them valuable for virtual assistants and information retrieval systems.

4. **Text Summarization:** LLMs can summarize long documents or articles, providing concise representations of the main points.

5. **Sentiment Analysis:** LLMs can determine the sentiment of a piece of text, aiding in customer feedback analysis and market research.

6. **Named Entity Recognition (NER):** LLMs can identify and categorize entities such as names of people, organizations, and locations within text.

7. **Language Modeling:** LLMs can predict the likelihood of a word or sentence in a given context, which is essential for various NLP tasks.

8. **Text Classification:** LLMs are used for tasks like spam detection, sentiment classification, and topic categorization.

The above applications demonstrate the versatility and significance of LLMs across different domains, highlighting their potential for enhancing natural language understanding and processing.




## **The history of LLMs**

The history of LLMs can be traced back to the early days of AI, when researchers began to experiment with ways to train computers to understand and generate natural language. One of the earliest LLMs was the ELIZA program, which was developed in the 1960s. ELIZA was a simple chatbot that could simulate a conversation with a human user.

In the 1980s, researchers began to develop more sophisticated LLMs, such as the ARPANET Natural Language Understanding (NLU) system. The ARPANET NLU system was able to understand and generate complex sentences, and it could even answer questions about the world.

In recent years, there has been a renewed interest in LLMs, due to the availability of large datasets of text and code. These datasets have allowed researchers to train LLMs that are much larger and more powerful than the LLMs that were developed in the past.

### **The ethical considerations of LLMs**

LLMs raise a number of ethical considerations, such as:

* The potential for bias: LLMs are trained on large datasets of text, which may contain biases. This means that LLMs may be biased in their outputs.
* The potential for misuse: LLMs could be used to generate harmful content, such as hate speech or propaganda.
* The potential for privacy violations: LLMs could be used to collect and process personal data without users' knowledge or consent.

It is important to be aware of these ethical considerations when using LLMs. Developers should take steps to mitigate the risks of bias, misuse, and privacy violations.

## **How do Transformers work?**

In this section, we will take a high-level look at the architecture of Transformer models.

### **A bit of Transformer history**
Here are some reference points in the (short) history of Transformer models:

A brief chronology of Transformers models.
The Transformer architecture was introduced in June 2017. The focus of the original research was on translation tasks. This was followed by the introduction of several influential models, including:

June 2018: GPT, the first pretrained Transformer model, used for fine-tuning on various NLP tasks and obtained state-of-the-art results

October 2018: BERT, another large pretrained model, this one designed to produce better summaries of sentences (more on this in the next chapter!)

February 2019: GPT-2, an improved (and bigger) version of GPT that was not immediately publicly released due to ethical concerns

October 2019: DistilBERT, a distilled version of BERT that is 60% faster, 40% lighter in memory, and still retains 97% of BERT’s performance

October 2019: BART and T5, two large pretrained models using the same architecture as the original Transformer model (the first to do so)

May 2020, GPT-3, an even bigger version of GPT-2 that is able to perform well on a variety of tasks without the need for fine-tuning (called zero-shot learning)

This list is far from comprehensive, and is just meant to highlight a few of the different kinds of Transformer models. Broadly, they can be grouped into three categories:

GPT-like (also called auto-regressive Transformer models)
BERT-like (also called auto-encoding Transformer models)
BART/T5-like (also called sequence-to-sequence Transformer models)
We will dive into these families in more depth later on.

### **Transformers are language models**

All the Transformer models mentioned above (GPT, BERT, BART, T5, etc.) have been trained as language models. This means they have been trained on large amounts of raw text in a self-supervised fashion. Self-supervised learning is a type of training in which the objective is automatically computed from the inputs of the model. That means that humans are not needed to label the data!

This type of model develops a statistical understanding of the language it has been trained on, but it’s not very useful for specific practical tasks. Because of this, the general pretrained model then goes through a process called transfer learning. During this process, the model is fine-tuned in a supervised way — that is, using human-annotated labels — on a given task.

An example of a task is predicting the next word in a sentence having read the n previous words. This is called causal language modeling because the output depends on the past and present inputs, but not the future ones.

#### **Pretrained vs Finetuning models**
**Pretraining** is the act of training a model from scratch: the weights are randomly initialized, and the training starts without any prior knowledge.This pretraining is usually done on very large amounts of data. Therefore, it requires a very large corpus of data, and training can take up to several weeks.

**Fine-tuning**, on the other hand, is the training done after a model has been pretrained. To perform fine-tuning, you first acquire a pretrained language model, then perform additional training with a dataset specific to your task.

## **Transformers**





### **The model is primarily composed of two blocks:**

Encoder (left): The encoder receives an input and builds a representation of it (its features). This means that the model is optimized to acquire understanding from the input.
Decoder (right): The decoder uses the encoder’s representation (features) along with other inputs to generate a target sequence. This means that the model is optimized for generating outputs.

image.png

Each of these parts can be used independently, depending on the task:

**Encoder-only models**: Good for tasks that require understanding of the input, such as sentence classification and named entity recognition.
Decoder-only models: Good for generative tasks such as text generation.
Encoder-decoder models or sequence-to-sequence models: Good for generative tasks that require an input, such as translation or summarization.

**Encoder-decoder models** (also called sequence-to-sequence models) use both parts of the Transformer architecture. At each stage, the attention layers of the encoder can access all the words in the initial sentence, whereas the attention layers of the decoder can only access the words positioned before a given word in the input.

#**Chapter 5: Prompt engineering**

**Prompt engineering** is a powerful technique to customize the outputs of large language models (LLMs) from Hugging Face. Here's a breakdown of how you can achieve this:

1. Understand the Task and Desired Outcome:

  - Clearly define the task you want the LLM to perform. Is it writing different kinds of creative content, translating languages, or answering your questions in a specific format?
  - Determine the desired outcome for the prompt. What kind of response or generation are you expecting from the LLM?
2. Craft the Prompt:
  -  Instruction: Clearly state what the LLM should do. Use concise and specific language to guide the model towards the desired outcome.
  - Context (Optional): Provide relevant background information or examples to help the LLM understand the task better. This can be especially useful for complex tasks.

  **Here are some prompt engineering techniques to consider**:

  - Few-Shot Prompting: Provide a few examples demonstrating the desired format or style of the output. This helps the LLM learn the pattern and replicate it in its generation.

  - Chain-of-Thought Prompting: Break down complex tasks into smaller steps with prompts for each step. This can improve the coherence and reasoning in the LLM's response.

  - Template-Based Prompting: Use pre-defined templates with placeholders for specific information. This can be efficient for tasks with a consistent structure.

3. Refine and Iterate:

  - Test your prompts with the LLM and analyze the outputs.
If the outputs are not as expected, refine your prompts by adjusting the instructions, context, or trying different techniques.

**Prompting vs fine-tuning**

You can achieve great results by optimizing your prompts, however, you may still ponder whether fine-tuning a model would work better for your case. Here are some scenarios when fine-tuning a smaller model may be a preferred option:

- Your domain is wildly different from what LLMs were pre-trained on and extensive prompt optimization did not yield sufficient results.
- You need your model to work well in a low-resource language.
- You need the model to be trained on sensitive data that is under strict regulations.
- You have to use a small model due to cost, privacy, infrastructure or other limitations.

In all of the above examples, you will need to make sure that you either already have or can easily obtain a large enough domain-specific dataset at a reasonable cost to fine-tune a model. You will also need to have enough time and resources to fine-tune a model.

# 6. **Fine tuning llm models**

**Fine-tuning**, is a technique used to adapt a pre-trained model to a new task. Here's a breakdown of the process:

- Pre-train a Model: You start with a model that has already been trained on a large and general dataset. This model learns low-level features that are often applicable to various tasks. For example, a pre-trained model for image recognition might learn to identify edges, shapes, and basic objects.

- Choose a Target Task:  You define a new, specific task you want the model to perform. This task should be related to the pre-training task but with a different goal. For instance, you might want the image recognition model to now classify specific types of animals (cats, dogs, birds) instead of just recognizing general objects.

- Freeze Base Layers (Optional):  In some cases, you might choose to freeze the weights of the initial layers (often called the base layers) of the pre-trained model. These layers capture the general features learned from the large dataset, and freezing them prevents them from being modified during fine-tuning. This maintains the learned knowledge while focusing the adaptation on the later layers.

- Fine-tune the Model: You adjust the weights of the final layers (or all layers if not frozen) of the pre-trained model using your new, smaller dataset specific to the target task. This process is similar to training a model from scratch, but it leverages the pre-trained knowledge as a starting point, requiring less data and training time.

- Evaluate and Refine:  Once fine-tuning is complete, you evaluate the model's performance on the target task. If the performance isn't satisfactory, you might need to adjust hyperparameters, the amount of data used, or even the layers being fine-tuned.

### **bitsandbytes**

Quantization, refers to the process of converting a data representation from a higher precision format to a lower precision format. This is done with the goal of reducing the amount of memory required to store the data and potentially speeding up computations.Loading model in 4 bit precision which means numbers are represented using only 4 bits instead of the usual 32 bits (single-precision float) and may reduce accuracy but while reducing precision saves memory and can speed up computations, it also introduces a loss of information. The accuracy of the calculations might be slightly affected but will recover after finetuning

```python
# Activate 4-bit precision base model loading
use_4bit = True

# Compute dtype for 4-bit base models
bnb_4bit_compute_dtype = "float16"

# Quantization type (fp4 or nf4)
bnb_4bit_quant_type = "nf4"

# Activate nested quantization for 4-bit base models (double quantization)
use_nested_quant = False

compute_dtype = getattr(torch, bnb_4bit_compute_dtype)

bnb_config = BitsAndBytesConfig(
    load_in_4bit=use_4bit,
    bnb_4bit_quant_type=bnb_4bit_quant_type,
    bnb_4bit_compute_dtype=compute_dtype,
    bnb_4bit_use_double_quant=use_nested_quant,
)

# Check GPU compatibility with bfloat16
if compute_dtype == torch.float16 and use_4bit:
    major, _ = torch.cuda.get_device_capability()
    if major >= 8:
        print("=" * 80)
        print("Your GPU supports bfloat16: accelerate training with bf16=True")
        print("=" * 80)
```

### **QLoRA parameters**

QLoRA, in this context, likely refers to Quantized Low-Rank Adaptation, a technique used in models to improve their adaptability to new tasks or data.

LLMs have a massive number of parameters, making them computationally expensive to train and store.

**Qlora** quantizes(reduce precision) by changing weight's from
higher precision formats(32 bit) to (4 bit) this significantly reduces the model's size, allowing it to fit on less powerful hardware and be loaded faster.

**Qlora** also uses low-rank component and learnable scaling factors to the model's architecture. This allows the model to adapt to new tasks with minimal changes to the original LLM parameters, reducing the amount of data needed for fine-tuning.


```python
# LoRA attention dimension
lora_r = 64

# Alpha parameter for LoRA scaling
lora_alpha = 16

# Dropout probability for LoRA layers
lora_dropout = 0.1

# Load LoRA configuration
peft_config = LoraConfig(
    lora_alpha=lora_alpha,
    lora_dropout=lora_dropout,
    r=lora_r,
    bias="none",
    task_type="CAUSAL_LM",
)
```

#6. **Evaluation of trained model.**


Evaluating large language models (LLMs) requires a comprehensive approach that considers various aspects to gauge their effectiveness and capabilities. Firstly, assessing the model's performance in tasks like natural language understanding (NLU) and generation is crucial. This involves testing its accuracy in answering questions, completing sentences, and generating coherent text.

Next, evaluating the model's robustness is essential. Robustness refers to its ability to handle diverse inputs, including noisy or adversarial data, without compromising performance. Conducting stress tests by introducing perturbations or challenging scenarios can reveal the model's resilience.

Furthermore, examining the model's ethical considerations is paramount. This includes evaluating biases in the data it was trained on and ensuring fairness and inclusivity in its outputs. Assessing the model's impact on societal issues, such as misinformation or stereotype reinforcement, is also crucial.

Scalability is another key aspect to evaluate, especially for LLMs designed for large-scale applications. Testing the model's efficiency in processing large volumes of data and its ability to scale horizontally or vertically is essential for real-world deployment.

Lastly, gathering feedback from domain experts and end-users through user studies or surveys can provide valuable insights into the model's usability, relevance, and overall performance in specific contexts. Incorporating these multi-faceted evaluation strategies can lead to a more comprehensive understanding of LLMs and their potential impact.


There are several techniques used to evaluate large language models (LLMs) effectively:

1. **Task-specific evaluation:** Assessing the model's performance on specific tasks such as text completion, sentiment analysis, summarization, or question-answering. This involves using benchmark datasets and metrics tailored to each task to measure accuracy and efficiency.

2. **Human evaluation:** Involves human annotators or judges assessing the quality of the model's outputs. This can include evaluating the fluency, coherence, relevance, and overall quality of generated text compared to human-generated text.

3. **Automatic evaluation metrics:** Utilizing automated metrics such as BLEU (Bilingual Evaluation Understudy), ROUGE (Recall-Oriented Understudy for Gisting Evaluation), METEOR (Metric for Evaluation of Translation with Explicit Ordering), and others. These metrics are commonly used in machine translation, text summarization, and other natural language processing tasks to quantify the similarity between the model's output and reference text.

4. **Adversarial testing:** Testing the model's robustness by introducing adversarial examples or perturbations. This helps evaluate its ability to handle noisy or misleading inputs without significantly compromising performance.

5. **Bias and fairness evaluation:** Assessing the model for biases in its training data and outputs. Techniques include measuring demographic biases, stereotype reinforcement, and fairness in decision-making processes to ensure equitable and unbiased performance.

6. **Scalability testing:** Evaluating the model's scalability by measuring its performance on large-scale datasets and assessing its ability to handle increasing computational demands as data volume grows.

7. **Ethical considerations:** Considering the ethical implications of using LLMs, including privacy concerns, potential misuse, and societal impact. Ethical evaluation frameworks help ensure responsible development and deployment of LLMs.

By combining these techniques, researchers and developers can gain a comprehensive understanding of an LLM's performance, robustness, fairness, scalability, and ethical implications, leading to more informed decisions and improvements in model development.

The most popular techniques used for evaluating Large Language Models (LLMs) include:

**GLUE** (General Language Understanding Evaluation) and SuperGLUE:
  - GLUE and SuperGLUE are widely used benchmarks for evaluating LLMs across a diverse set of tasks such as sentiment analysis, natural language inference, paraphrase detection, and more. They provide standardized tasks and evaluation metrics, allowing for fair comparisons between different models.

  - **When to Use:** GLUE and SuperGLUE are suitable for evaluating the overall language understanding capabilities of LLMs across a wide range of tasks. These benchmarks are ideal for comparative analysis between different models and for assessing generalization and performance across diverse NLP tasks. Use GLUE and SuperGLUE when you need a standardized benchmark to compare the performance of LLMs in tasks like text classification, sentiment analysis, natural language inference, and semantic similarity.

**BLEU** (Bilingual Evaluation Understudy):
- BLEU is a metric commonly used to evaluate the quality of machine-generated text, particularly in machine translation tasks. It measures the similarity between machine-generated text and human-generated reference text based on n-gram precision.
- **When to Use BLEU:** BLEU is particularly useful for evaluating the quality of machine-generated text in tasks such as machine translation. Use BLEU when you need to measure the similarity between machine-generated text and human-generated reference text based on n-gram precision. It's commonly used in machine translation evaluations.

**ROUGE** (Recall-Oriented Understudy for Gisting Evaluation):
- ROUGE is another metric used for evaluating the quality of machine-generated text, focusing on text summarization tasks. It measures the overlap between machine-generated summaries and human-generated reference summaries.
-    **When to Use ROUGE:** ROUGE is beneficial for evaluating the quality of machine-generated summaries in text summarization tasks. Use ROUGE when you need to measure the overlap between machine-generated summaries and human-generated reference summaries. It's commonly used in automatic summarization evaluations.

**Perplexity**: Perplexity is a metric often used to evaluate the language model's ability to predict the next word in a sequence. Lower perplexity values indicate better performance, as the model can predict the next word more accurately.
   - **When to Use:** Perplexity is often used to evaluate the language model's ability to predict the next word in a sequence. It's commonly used in tasks like language modeling and text generation. Use perplexity when you need to assess the fluency and predictive accuracy of an LLM by measuring how well it predicts the next word in a given context. Lower perplexity values indicate better predictive performance.

In summary, choose GLUE and SuperGLUE for comprehensive evaluation across diverse NLP tasks, BLEU for machine translation quality assessment, ROUGE for text summarization quality assessment, perplexity for language modeling and text generation evaluation, and F1 score for tasks like NER and POS tagging that require a balance between precision and recall. Each technique serves specific evaluation needs based on the task and metrics of interest.
